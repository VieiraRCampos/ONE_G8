#üìå Prepara√ß√£o dos Dados

```python
# Importar bibliotecas necess√°rias
import pandas as pd
import requests
import json
from IPython.display import display

# URL do arquivo JSON no GitHub
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"

# Fazer requisi√ß√£o para a API
try:
    response = requests.get(url)

    # Verificar se a requisi√ß√£o foi bem sucedida
    if response.status_code == 200:
        print("‚úÖ Dados carregados com sucesso!")

        # Carregar os dados JSON
        dados_json = response.json()

        # Verificar a estrutura dos dados
        print(f"Tipo de dados: {type(dados_json)}")

        # Se for uma lista, podemos converter diretamente para DataFrame
        if isinstance(dados_json, list):
            df = pd.DataFrame(dados_json)
            print(f"DataFrame criado com {len(df)} linhas e {len(df.columns)} colunas")

            # Mostrar primeiras linhas
            print("\nüìä Primeiras 5 linhas do DataFrame:")
            display(df.head())

            # Informa√ß√µes b√°sicas do DataFrame
            print("\nüìà Informa√ß√µes do DataFrame:")
            print(df.info())

            # Estat√≠sticas descritivas
            print("\nüìä Estat√≠sticas descritivas:")
            display(df.describe())

        else:
            print("Estrutura de dados n√£o √© uma lista. Verificando estrutura...")
            print(f"Tipo: {type(dados_json)}")
            if isinstance(dados_json, dict):
                print("Chaves dispon√≠veis:", dados_json.keys())

except Exception as e:
    print(f"‚ùå Erro ao carregar dados: {e}")
```
