#📌 Preparação dos Dados

```python
# Importar bibliotecas necessárias
import pandas as pd
import requests
import json
from IPython.display import display

# URL do arquivo JSON no GitHub
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"

# Fazer requisição para a API
try:
    response = requests.get(url)

    # Verificar se a requisição foi bem sucedida
    if response.status_code == 200:
        print("✅ Dados carregados com sucesso!")

        # Carregar os dados JSON
        dados_json = response.json()

        # Verificar a estrutura dos dados
        print(f"Tipo de dados: {type(dados_json)}")

        # Se for uma lista, podemos converter diretamente para DataFrame
        if isinstance(dados_json, list):
            df = pd.DataFrame(dados_json)
            print(f"DataFrame criado com {len(df)} linhas e {len(df.columns)} colunas")

            # Mostrar primeiras linhas
            print("\n📊 Primeiras 5 linhas do DataFrame:")
            display(df.head())

            # Informações básicas do DataFrame
            print("\n📈 Informações do DataFrame:")
            print(df.info())

            # Estatísticas descritivas
            print("\n📊 Estatísticas descritivas:")
            display(df.describe())

        else:
            print("Estrutura de dados não é uma lista. Verificando estrutura...")
            print(f"Tipo: {type(dados_json)}")
            if isinstance(dados_json, dict):
                print("Chaves disponíveis:", dados_json.keys())

except Exception as e:
    print(f"❌ Erro ao carregar dados: {e}")
```
