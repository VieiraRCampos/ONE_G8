````python
# =============================================================================
# 🛠️ TRATAMENTO DE DADOS FALTANTES E INCONSISTENTES
# =============================================================================

print("="*80)
print("🛠️  INICIANDO TRATAMENTO DE DADOS FALTANTES")
print("="*80)

# Import numpy for handling NaN values
import numpy as np
import pandas as pd

# Fazer uma cópia do DataFrame original para preservar os dados brutos
# df_clean = df.copy() # No longer needed as we will create df_transformed

# Extract data from nested dictionary columns first
customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])

# Concatenate the new dataframes with the original dataframe (excluding original nested columns)
df_transformed = pd.concat([df[['customerID', 'Churn']], customer_df, phone_df, internet_df, account_df], axis=1)

# Now work with df_transformed for cleaning and analysis
df_clean = df_transformed.copy() # Use df_transformed as the base for cleaning
print(f"✅ DataFrame para limpeza criado (df_clean): {df_clean.shape}")


# Convert potential numeric columns to numeric, coercing errors
# This was already done in YTicbhNefyAE, repeating here for completeness in this cell
for col in ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges']:
    if col in df_clean.columns:
        # Replace empty strings and spaces with NaN before converting
        df_clean[col] = df_clean[col].replace(r'^\s*$', np.nan, regex=True)
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')


# =============================================================================
# 1. ANÁLISE COMPLETA DOS DADOS FALTANTES
# =============================================================================

def analise_dados_faltantes(df):
    """Analisa dados faltantes de forma abrangente"""
    print("\n" + "="*60)
    print("🔍 ANÁLISE DETALHADA DOS DADOS FALTANTES")
    print("="*60)

    missing_analysis = pd.DataFrame(index=df.columns)

    # Valores nulos tradicionais
    missing_analysis['Nulos'] = df.isnull().sum()
    missing_analysis['Percentual_Nulos'] = (df.isnull().sum() / len(df)) * 100

    # Strings vazias - Check only for object columns after numeric conversion
    missing_analysis['Strings_Vazias'] = 0
    for col in df.columns:
        if df[col].dtype == 'object':
            missing_analysis.loc[col, 'Strings_Vazias'] = (df[col] == "").sum()

    # Apenas espaços em branco - Check only for object columns after numeric conversion
    missing_analysis['Apenas_Espacos'] = 0
    for col in df.columns:
        if df[col].dtype == 'object':
            missing_analysis.loc[col, 'Apenas_Espacos'] = df[col].apply(
                lambda x: isinstance(x, str) and x.strip() == ""
            ).sum()

    # Total de dados faltantes
    missing_analysis['Total_Faltantes'] = (
        missing_analysis['Nulos'] +
        missing_analysis['Strings_Vazias'] +
        missing_analysis['Apenas_Espacos']
    )

    missing_analysis['Percentual_Total'] = (missing_analysis['Total_Faltantes'] / len(df)) * 100

    # Mostrar apenas colunas com dados faltantes
    missing_data = missing_analysis[missing_analysis['Total_Faltantes'] > 0]

    if len(missing_data) > 0:
        print("📊 DADOS FALTANTES ENCONTRADOS:")
        display(missing_data.sort_values('Total_Faltantes', ascending=False))
        return missing_data
    else:
        print("✅ Nenhum dado faltante encontrado!")
        return None

# Executar análise no df_clean
missing_data = analise_dados_faltantes(df_clean)

# =============================================================================
# 2. ESTRATÉGIAS DE TRATAMENTO POR TIPO DE COLUNA
# =============================================================================

print("\n" + "="*60)
print("🎯 APLICANDO ESTRATÉGIAS DE TRATAMENTO")
print("="*60)

# Dicionário para armazenar estatísticas do tratamento
tratamento_stats = {}

# Helper functions for treatment (already defined, keeping them here)
def tratar_coluna_numerica(df, coluna, estrategia='media'):
    """Trata dados faltantes em colunas numéricas"""
    antes = df[coluna].isnull().sum()

    if estrategia == 'media':
        valor_substituicao = df[coluna].mean()
        df[coluna] = df[coluna].fillna(valor_substituicao)
        metodo = f'média ({valor_substituicao:.2f})'
    elif estrategia == 'mediana':
        valor_substituicao = df[coluna].median()
        df[coluna] = df[coluna].fillna(valor_substituicao)
        metodo = f'mediana ({valor_substituicao:.2f})'
    elif estrategia == 'zero':
        df[coluna] = df[coluna].fillna(0)
        metodo = 'zero'
    elif estrategia == 'moda': # Moda para numéricos pode ser útil
         moda = df[coluna].mode()
         if not moda.empty:
             valor_substituicao = moda[0]
             df[coluna] = df[coluna].fillna(valor_substituicao)
             metodo = f'moda ({valor_substituicao})'
         else:
             df[coluna] = df[coluna].fillna(df[coluna].mean()) # Fallback to mean if no mode
             metodo = 'média (sem moda)'


    depois = df[coluna].isnull().sum()
    tratados = antes - depois

    return tratados, metodo

def tratar_coluna_categorica(df, coluna, estrategia='moda'):
    """Trata dados faltantes em colunas categóricas"""
    antes = df[coluna].isnull().sum() + (df[coluna] == "").sum()

    if estrategia == 'moda':
        # Encontrar a moda (valor mais frequente), excluindo vazios e nulos para calcular a moda
        moda = df[coluna][(df[coluna] != "") & (df[coluna].notna())].mode()
        if not moda.empty:
            valor_substituicao = moda[0]
            df[coluna] = df[coluna].replace('', valor_substituicao).fillna(valor_substituicao)
            metodo = f'moda ({valor_substituicao})'
        else:
             # If mode is empty (e.g., all values are missing), use "Desconhecido"
            df[coluna] = df[coluna].replace('', 'Desconhecido').fillna('Desconhecido')
            metodo = '"Desconhecido"'
    elif estrategia == 'desconhecido':
        df[coluna] = df[coluna].replace('', 'Desconhecido').fillna('Desconhecido')
        metodo = '"Desconhecido"'
    elif estrategia == 'vazio':
        df[coluna] = df[coluna].fillna('')
        metodo = 'string vazia'

    depois = df[coluna].isnull().sum() + (df[coluna] == "").sum()
    tratados = antes - depois

    return tratados, metodo

# =============================================================================
# 3. APLICAR TRATAMENTO ESPECÍFICO POR COLUNA
# =============================================================================

if missing_data is not None:
    for coluna in missing_data.index:
        total_faltantes = missing_data.loc[coluna, 'Total_Faltantes']

        if total_faltantes > 0:
            print(f"\n📌 Tratando coluna: {coluna}")
            print(f"   Dados faltantes: {total_faltantes}")

            # Definir estratégia baseada no tipo de dados e contexto
            if df_clean[coluna].dtype in ['int64', 'float64']:
                # Colunas numéricas
                if 'charge' in coluna.lower() or 'amount' in coluna.lower() or 'price' in coluna.lower():
                    tratados, metodo = tratar_coluna_numerica(df_clean, coluna, 'zero')
                else:
                    tratados, metodo = tratar_coluna_numerica(df_clean, coluna, 'media')

                print(f"   ✅ Numérico - Substituídos {tratados} valores por {metodo}")

            else:
                # Colunas categóricas
                if 'churn' in coluna.lower():
                    # Para variável de evasão, análise especial
                    print(f"   ⚠️  Variável de evasão - análise manual recomendada")

                    # Mostrar distribuição antes do tratamento
                    print(f"   Distribuição antes do tratamento:")
                    print(df_clean[coluna].value_counts(dropna=False))

                    # Decisão: substituir pela moda (mais seguro no início)
                    tratados, metodo = tratar_coluna_categorica(df_clean, coluna, 'moda')
                    print(f"   ✅ Substituídos {tratados} valores por {metodo}")

                else:
                    tratados, metodo = tratar_coluna_categorica(df_clean, coluna, 'moda')
                    print(f"   ✅ Categórico - Substituídos {tratados} valores por {metodo}")

            # Armazenar estatísticas
            tratamento_stats[coluna] = {
                'antes': total_faltantes,
                'tratados': tratados,
                'metodo': metodo
            }

# =============================================================================
# 4. TRATAMENTO ESPECIAL PARA A VARIÁVEL DE EVASÃO
# =============================================================================

print("\n" + "="*60)
print("🎯 TRATAMENTO ESPECIAL - VARIÁVEL DE EVASÃO")
print("="*60)

# Identificar coluna de evasão
churn_cols = [col for col in df_clean.columns if 'churn' in col.lower()]
if churn_cols:
    churn_col = churn_cols[0]
    print(f"Analisando variável de evasão: {churn_col}")

    # Verificar se ainda há dados problemáticos (strings vazias ou nulos)
    churn_problems = (df_clean[churn_col].isnull().sum() +
                     (df_clean[churn_col] == "").sum())

    if churn_problems > 0:
        print(f"⚠️  Ainda existem {churn_problems} valores problemáticos em {churn_col}")

        # Análise detalhada dos valores
        print("Distribuição atual:")
        print(df_clean[churn_col].value_counts(dropna=False))

        # Estratégia conservadora: para evasão, considerar como "No" se não especificado
        # (assumindo que a maioria dos clientes não dá churn)
        # Only apply if the column is object type and contains 'Yes', 'No', '', or None
        if df_clean[churn_col].dtype == 'object':
             # Check if unique values are a subset of expected churn values
            if set(df_clean[churn_col].unique()).issubset({'Yes', 'No', '', None}):
                print("💡 Convertendo valores faltantes para 'No'")
                df_clean[churn_col] = df_clean[churn_col].replace('', 'No').fillna('No')

                print("Distribuição após tratamento:")
                print(df_clean[churn_col].value_counts())
            else:
                print(f"ℹ️  Coluna '{churn_col}' contém valores inesperados, tratamento manual recomendado.")


# =============================================================================
# 5. VERIFICAÇÃO FINAL DA QUALIDADE DOS DADOS
# =============================================================================

print("\n" + "="*60)
print("✅ VERIFICAÇÃO FINAL DA QUALIDADE DOS DADOS")
print("="*60)

# Verificar se ainda existem dados faltantes (nulos ou strings vazias em colunas object)
missing_final_nulos = df_clean.isnull().sum().sum()
missing_final_vazias = sum((df_clean[col] == "").sum() for col in df_clean.columns if df_clean[col].dtype == 'object')
missing_final = missing_final_nulos + missing_final_vazias


if missing_final == 0:
    print("🎉 TODOS OS DADOS FALTANTES FORAM TRATADOS COM SUCESSO!")
else:
    print(f"⚠️  Ainda existem {missing_final} dados faltantes no dataset")

    # Detalhar os remanescentes
    remaining_missing = pd.DataFrame({
        'Nulos': df_clean.isnull().sum(),
        'Strings_Vazias': (df_clean.select_dtypes(include='object') == "").sum() # Only check object columns for empty strings
    })
    remaining_missing = remaining_missing[(remaining_missing['Nulos'] > 0) |
                                        (remaining_missing['Strings_Vazias'] > 0)]

    if len(remaining_missing) > 0:
        print("Dados faltantes remanescentes:")
        display(remaining_missing)

# =============================================================================
# 6. ESTATÍSTICAS DO TRATAMENTO APLICADO
# =============================================================================

print("\n" + "="*60)
print("📊 ESTATÍSTICAS DO TRATAMENTO APLICADO")
print("="*60)

if tratamento_stats:
    total_tratados = sum(stats['tratados'] for stats in tratamento_stats.values())
    print(f"Total de valores tratados: {total_tratados}")

    print("\nDetalhamento por coluna:")
    for coluna, stats in tratamento_stats.items():
        print(f"   {coluna}: {stats['tratados']} valores → {stats['metodo']}")
else:
    print("Nenhum tratamento foi necessário")

# =============================================================================
# 7. CONVERSÃO DE TIPOS DE DADOS (OPCIONAL)
# =============================================================================

print("\n" + "="*60)
print("🔄 OTIMIZAÇÃO DE TIPOS DE DADOS")
print("="*60)

# Converter colunas categóricas para tipo category para economizar memória
# Apply this on df_clean after flattening
categorical_cols = df_clean.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df_clean[col].nunique() < 50:  # Only if has few unique values
        df_clean[col] = df_clean[col].astype('category')
        print(f"✅ {col} convertido para category")
    else:
         print(f"ℹ️  {col} não convertido para category (muitos valores únicos)")


print(f"\n🎯 DATASET FINAL PRONTO PARA ANÁLISE!")
print(f"   Dimensões: {df_clean.shape}")
print(f"   Memória utilizada: {df_clean.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# Mostrar amostra dos dados tratados
print("\n📋 AMOSTRA DOS DADOS TRATADOS:")
display(df_clean.head(3))
````

````python
# =============================================================================
# 🎯 CRIAR CONTAS_DIARIAS
# =============================================================================

print("="*50)
print("💰 CRIANDO COLUNA CONTAS_DIARIAS (SIMPLIFICADO)")
print("="*50)

# Supondo que a coluna de faturamento mensal se chama 'MonthlyCharges'
coluna_mensal = 'Charges.Monthly'

if coluna_mensal in df_clean.columns:
    # Calcular conta diária (30 dias por mês)
    df_clean['Contas_Diarias'] = df_clean[coluna_mensal] / 30

    print(f"✅ Coluna criada com sucesso!")
    print(f"   Fonte: {coluna_mensal}")
    print(f"   Média das contas diárias: {df_clean['Contas_Diarias'].mean():.2f}")
    print(f"   Valores entre {df_clean['Contas_Diarias'].min():.2f} e {df_clean['Contas_Diarias'].max():.2f}")

    # Mostrar amostra
    print(f"\n📋 Amostra:")
    display(df_clean[[coluna_mensal, 'Contas_Diarias']].head())

else:
    print(f"❌ Coluna '{coluna_mensal}' não encontrada")
    print("📋 Colunas disponíveis:")
    for col in df_clean.columns:
        print(f"   - {col}")
````

````python
# =============================================================================
# 🎯 VERSÃO CONVERSÃO BINÁRIA
# =============================================================================

print("="*60)
print("🔢 CONVERSÃO SIMPLES SIM/NÃO → 1/0")
print("="*60)

# Mapeamento básico
binary_map_simple = {'Yes': 1, 'No': 0, 'Sim': 1, 'Não': 0, True: 1, False: 0}

# Colunas comuns em datasets de telecom
common_binary_cols = [
    'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling',
    'SeniorCitizen', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Aplicar conversão apenas para colunas existentes
cols_to_convert = [col for col in common_binary_cols if col in df_clean.columns]

print("Colunas para conversão binária:")
for col in cols_to_convert:
    print(f"   🔄 {col}: {df_clean[col].unique()}")

# Apply conversion
for col in cols_to_convert:
    # Convert to object dtype first to avoid issues with categorical operations
    df_clean[col] = df_clean[col].astype('object')

    # Handle 'No internet service' specifically before mapping
    if 'No internet service' in df_clean[col].unique():
        df_clean[col] = df_clean[col].replace('No internet service', 'No')

    # Apply the binary map and handle potential NaNs if any were introduced
    # Ensure the result is numeric
    df_clean[col] = df_clean[col].map(binary_map_simple).fillna(df_clean[col])
    # After mapping, ensure the column is numeric if it contains only 0s and 1s
    if set(df_clean[col].dropna().unique()).issubset({0, 1}):
         df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')


    print(f"✅ {col}: convertida")

print("\n📋 Amostra após conversão:")
display(df_clean[cols_to_convert].head())
````

````python
# =============================================================================
# 📊 ANÁLISE DESCRITIVA COMPLETA DO DATAFRAME
# =============================================================================

print("="*80)
print("📈 ANÁLISE DESCRITIVA COMPLETA - TELECOM CLIENTES")
print("="*80)

print(f"📋 Dimensões do Dataset: {df_clean.shape[0]} linhas × {df_clean.shape[1]} colunas")
print(f"💾 Memória utilizada: {df_clean.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# =============================================================================
# 1. ANÁLISE GERAL DOS TIPOS DE DADOS
# =============================================================================

print("\n" + "="*60)
print("🔍 DISTRIBUIÇÃO DOS TIPOS DE DADOS")
print("="*60)

# Contagem por tipo de dado
type_counts = df_clean.dtypes.value_counts()
for dtype, count in type_counts.items():
    print(f"   {dtype}: {count} colunas")

# Listar colunas por tipo
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df_clean.select_dtypes(include=['object', 'category']).columns.tolist()
bool_cols = df_clean.select_dtypes(include=['bool']).columns.tolist()

print(f"\n   🔢 Numéricas: {len(numeric_cols)} colunas")
print(f"   🏷️  Categóricas: {len(categorical_cols)} colunas")
print(f"   ⚡ Booleanas: {len(bool_cols)} colunas")

# =============================================================================
# 2. ANÁLISE DAS VARIÁVEIS NUMÉRICAS
# =============================================================================

print("\n" + "="*60)
print("📊 ESTATÍSTICAS DESCRITIVAS - VARIÁVEIS NUMÉRICAS")
print("="*60)

if numeric_cols:
    # Estatísticas descritivas básicas
    desc_numeric = df_clean[numeric_cols].describe().T
    desc_numeric['cv'] = (desc_numeric['std'] / desc_numeric['mean']).round(3)  # Coeficiente de variação
    desc_numeric['range'] = desc_numeric['max'] - desc_numeric['min']
    desc_numeric['iqr'] = desc_numeric['75%'] - desc_numeric['25%']

    print("Estatísticas descritivas das variáveis numéricas:")
    display(desc_numeric.round(3))

    # Análise de assimetria (skewness) e curtose
    print("\n📈 ANÁLISE DE ASSIMETRIA E CURTOSE:")
    skew_kurtosis = pd.DataFrame({
        'Skewness': df_clean[numeric_cols].skew().round(3),
        'Kurtosis': df_clean[numeric_cols].kurtosis().round(3)
    })

    # Classificar a assimetria
    skew_kurtosis['Tipo_Assimetria'] = skew_kurtosis['Skewness'].apply(
        lambda x: 'Simétrica' if abs(x) < 0.5 else
                 'Moderada' if abs(x) < 1 else
                 'Forte'
    )

    display(skew_kurtosis)

    # Identificar outliers usando IQR
    print("\n🔍 DETECÇÃO DE OUTLIERS (Método IQR):")
    outliers_info = {}
    for col in numeric_cols:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df_clean[(df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)]
        outlier_count = len(outliers)
        outlier_percent = (outlier_count / len(df_clean)) * 100

        outliers_info[col] = {
            'Outliers': outlier_count,
            'Percentual': f"{outlier_percent:.2f}%",
            'Limite_Inferior': lower_bound,
            'Limite_Superior': upper_bound
        }

    outliers_df = pd.DataFrame(outliers_info).T
    print("Quantidade de outliers por variável:")
    display(outliers_df[['Outliers', 'Percentual']].sort_values('Outliers', ascending=False))

else:
    print("❌ Nenhuma variável numérica encontrada")

# =============================================================================
# 3. ANÁLISE DAS VARIÁVEIS CATEGÓRICAS
# =============================================================================

print("\n" + "="*60)
print("🏷️  ANÁLISE DAS VARIÁVEIS CATEGÓRICAS")
print("="*60)

if categorical_cols:
    print("📋 RESUMO DAS VARIÁVEIS CATEGÓRICAS:")
    cat_summary = pd.DataFrame(index=categorical_cols)
    cat_summary['Valores_Únicos'] = [df_clean[col].nunique() for col in categorical_cols]
    cat_summary['Valores_Faltantes'] = [df_clean[col].isnull().sum() for col in categorical_cols]
    cat_summary['Moda'] = [df_clean[col].mode()[0] if not df_clean[col].mode().empty else None for col in categorical_cols]
    cat_summary['Frequência_Moda'] = [df_clean[col].value_counts().iloc[0] if not df_clean[col].value_counts().empty else 0 for col in categorical_cols]
    cat_summary['Percentual_Moda'] = [(df_clean[col].value_counts().iloc[0] / len(df_clean) * 100).round(2) if not df_clean[col].value_counts().empty else 0 for col in categorical_cols]

    display(cat_summary.sort_values('Valores_Únicos'))

    # Análise detalhada para cada variável categórica
    print("\n📊 DISTRIBUIÇÃO DETALHADA POR VARIÁVEL CATEGÓRICA:")
    for col in categorical_cols:
        print(f"\n🎯 {col}:")
        value_counts = df_clean[col].value_counts()
        value_percent = df_clean[col].value_counts(normalize=True) * 100

        dist_df = pd.DataFrame({
            'Frequência': value_counts,
            'Percentual': value_percent.round(2)
        })
        display(dist_df.head(10))  # Mostrar as 10 categorias mais frequentes

        if len(value_counts) > 10:
            print(f"... e mais {len(value_counts) - 10} categorias")

else:
    print("✅ Nenhuma variável categórica encontrada")

# =============================================================================
# 4. ANÁLISE DA VARIÁVEL DE EVASÃO (CHURN)
# =============================================================================

print("\n" + "="*60)
print("🎯 ANÁLISE DA VARIÁVEL DE EVASÃO (CHURN)")
print("="*60)

# Encontrar coluna de churn
churn_cols = [col for col in df_clean.columns if 'churn' in col.lower()]
if churn_cols:
    churn_col = churn_cols[0]
    print(f"Variável de evasão identificada: '{churn_col}'")

    # Distribuição geral
    churn_distribution = df_clean[churn_col].value_counts()
    churn_percentage = df_clean[churn_col].value_counts(normalize=True) * 100

    print("\n📈 DISTRIBUIÇÃO GERAL:")
    for value, count in churn_distribution.items():
        perc = churn_percentage[value]
        print(f"   {value}: {count} clientes ({perc:.2f}%)")

    # Taxa de evasão geral
    if churn_col in numeric_cols:
        churn_rate = df_clean[churn_col].mean() * 100
        print(f"\n📊 Taxa de evasão geral: {churn_rate:.2f}%")
    else:
        # Se for categórica, assumir que 'Yes'/'Sim' representa churn
        churn_values = [v for v in churn_distribution.index if str(v).lower() in ['yes', 'sim', '1', 'true']]
        if churn_values:
            churn_count = sum(churn_distribution[v] for v in churn_values)
            churn_rate = (churn_count / len(df_clean)) * 100
            print(f"\n📊 Taxa de evasão geral: {churn_rate:.2f}%")

else:
    print("❌ Variável de evasão não encontrada")

# =============================================================================
# 5. ANÁLISE DE CORRELAÇÕES
# =============================================================================

print("\n" + "="*60)
print("🔗 MATRIZ DE CORRELAÇÕES")
print("="*60)

if len(numeric_cols) > 1:
    # Calcular matriz de correlação
    correlation_matrix = df_clean[numeric_cols].corr()

    print("Matriz de correlação entre variáveis numéricas:")
    display(correlation_matrix.round(3))

    # Encontrar correlações fortes (abs > 0.5)
    strong_correlations = []
    for i in range(len(correlation_matrix.columns)):
        for j in range(i+1, len(correlation_matrix.columns)):
            corr = correlation_matrix.iloc[i, j]
            if abs(corr) > 0.5:
                strong_correlations.append({
                    'Variável 1': correlation_matrix.columns[i],
                    'Variável 2': correlation_matrix.columns[j],
                    'Correlação': corr.round(3)
                })

    if strong_correlations:
        print("\n📊 CORRELAÇÕES FORTES (|r| > 0.5):")
        strong_corr_df = pd.DataFrame(strong_correlations)
        display(strong_corr_df.sort_values('Correlação', key=abs, ascending=False))
    else:
        print("ℹ️  Nenhuma correlação forte encontrada (|r| > 0.5)")

else:
    print("ℹ️  Não há variáveis numéricas suficientes para análise de correlação")

# =============================================================================
# 6. ANÁLISE POR GRUPOS (SE HOUVER VARIÁVEL DE EVASÃO)
# =============================================================================

print("\n" + "="*60)
print("👥 ANÁLISE COMPARATIVA POR GRUPOS")
print("="*60)

if churn_cols and churn_col in df_clean.columns:
    print(f"📊 COMPARAÇÃO ENTRE CLIENTES QUE DERAM CHURN vs NÃO-CHURN:")

    # Verificar se a coluna é numérica ou categórica
    if df_clean[churn_col].dtype in ['int64', 'float64']:
        # Assumir que 1 = churn, 0 = não-churn
        churn_group = df_clean[df_clean[churn_col] == 1]
        non_churn_group = df_clean[df_clean[churn_col] == 0]
    else:
        # Tentar identificar valores de churn
        churn_values = [v for v in df_clean[churn_col].unique()
                       if str(v).lower() in ['yes', 'sim', '1', 'true']]
        non_churn_values = [v for v in df_clean[churn_col].unique()
                           if str(v).lower() in ['no', 'não', '0', 'false']]

        if churn_values and non_churn_values:
            churn_group = df_clean[df_clean[churn_col].isin(churn_values)]
            non_churn_group = df_clean[df_clean[churn_col].isin(non_churn_values)]
        else:
            churn_group = non_churn_group = pd.DataFrame()
            print("❌ Não foi possível identificar grupos de churn")

    if len(churn_group) > 0 and len(non_churn_group) > 0:
        print(f"   Churn: {len(churn_group)} clientes")
        print(f"   Não-Churn: {len(non_churn_group)} clientes")

        # Comparar médias das variáveis numéricas
        print("\n📈 COMPARAÇÃO DE MÉDIAS:")
        comparison_data = []
        for col in numeric_cols:
            if col != churn_col:
                mean_churn = churn_group[col].mean()
                mean_non_churn = non_churn_group[col].mean()
                diff = mean_churn - mean_non_churn
                diff_percent = (diff / mean_non_churn * 100) if mean_non_churn != 0 else 0

                comparison_data.append({
                    'Variável': col,
                    'Média_Churn': mean_churn,
                    'Média_Não_Churn': mean_non_churn,
                    'Diferença': diff,
                    'Diferença_%': diff_percent
                })

        comparison_df = pd.DataFrame(comparison_data)
        comparison_df = comparison_df.sort_values('Diferença', key=abs, ascending=False)
        display(comparison_df.round(3).head(10))  # Mostrar as 10 maiores diferenças

# =============================================================================
# 7. ANÁLISE DE VALORES FALTANTES
# =============================================================================

print("\n" + "="*60)
print("🔍 ANÁLISE DE VALORES FALTANTES")
print("="*60)

# Verificar valores nulos
null_analysis = df_clean.isnull().sum()
null_percentage = (df_clean.isnull().sum() / len(df_clean)) * 100

null_df = pd.DataFrame({
    'Valores_Nulos': null_analysis,
    'Percentual_Nulos': null_percentage.round(2)
})

null_df = null_df[null_df['Valores_Nulos'] > 0].sort_values('Percentual_Nulos', ascending=False)

if len(null_df) > 0:
    print("⚠️  VALORES NULOS ENCONTRADOS:")
    display(null_df)
else:
    print("✅ Nenhum valor nulo encontrado no dataset")

# =============================================================================
# 8. VISUALIZAÇÕES RÁPIDAS
# =============================================================================

print("\n" + "="*60)
print("📊 VISUALIZAÇÕES RÁPIDAS")
print("="*60)

import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo
plt.style.use('default')
sns.set_palette("husl")

# Criar visualizações para as principais variáveis numéricas
if numeric_cols:
    # Selecionar até 6 variáveis numéricas para visualização
    cols_to_plot = numeric_cols[:6] if len(numeric_cols) > 6 else numeric_cols

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()

    for i, col in enumerate(cols_to_plot):
        if i < len(axes):
            axes[i].hist(df_clean[col].dropna(), bins=30, alpha=0.7, color='skyblue', edgecolor='black')
            axes[i].set_title(f'Distribuição de {col}')
            axes[i].set_xlabel(col)
            axes[i].set_ylabel('Frequência')
            axes[i].grid(True, alpha=0.3)

    # Remover eixos extras se necessário
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# =============================================================================
# 9. RESUMO EXECUTIVO
# =============================================================================

print("\n" + "="*80)
print("🎯 RESUMO EXECUTIVO DA ANÁLISE DESCRITIVA")
print("="*80)

print("📈 PRINCIPAIS INSIGHTS:")
print(f"   • Total de clientes: {len(df_clean):,}")
print(f"   • Variáveis analisadas: {len(df_clean.columns)}")

if churn_cols:
    churn_rate = (df_clean[churn_col].mean() * 100) if df_clean[churn_col].dtype in ['int64', 'float64'] else \
                (df_clean[churn_col].isin(['Yes', 'Sim', '1', 'True']).mean() * 100)
    print(f"   • Taxa de evasão: {churn_rate:.2f}%")

if numeric_cols:
    print(f"   • Variáveis numéricas: {len(numeric_cols)}")
    print(f"   • Variável numérica com maior variação: {desc_numeric['std'].idxmax()} (std: {desc_numeric['std'].max():.2f})")

if categorical_cols:
    most_categories = cat_summary['Valores_Únicos'].idxmax()
    print(f"   • Variável categórica com mais categorias: {most_categories} ({cat_summary.loc[most_categories, 'Valores_Únicos']} categorias)")


print("\n✅ ANÁLISE DESCRITIVA CONCLUÍDA!")
````
